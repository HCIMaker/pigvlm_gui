```python
from moviepy.video.io.VideoFileClip import VideoFileClip
```

```python
def hms_to_secs(datetime:str) -> float:
    parts = [float(x) for x in datetime.strip().split(":")] # strip() removes whitespace
    if len(parts) == 1:
        return parts[0] # It should only have seconds
    if len(parts) == 2:
        return parts[0] * 60 + parts[1] # It should have minutes and seconds
    if len(parts) == 3:
        return parts[0] * 3600 + parts[1] * 60 + parts[2] # It should have hours, minutes and seconds
    raise ValueError(f"Bad time format: {datetime}")

input_video = r"D:\Research\PigVLM\PigFarmDataProcessing\projects\datasets\1_raw\Video\Crate_08\2022-04-28\ch07_Crate08_20220428120000_20220428140001.mp4"
output_video = r"./output.mp4"
start_time = "00:58:00"
end_time   = "00:59:00"
```

```python
with VideoFileClip(input_video) as clip:
    sub = clip.subclip(hms_to_secs(start_time), hms_to_secs(end_time))
    sub.write_videofile(output_video, codec="libx264", audio_codec="aac")
```
